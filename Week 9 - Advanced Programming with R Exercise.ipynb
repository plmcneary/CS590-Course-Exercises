1. Import peanut_lines.csv and store data as lines
2. Show the first 25 rows of lines

```r
peanut_lines <- read.csv("peanut_lines.csv")
peanut_lines[1:25,]
```

3. Count the number of lines that have Bailey as a parent using grep()

```r
length(grep("Bailey", peanut_lines$Parentage))
```

4. Create a new column called 'Bailey_Parent' where the value is 'Yes' if 'Bailey' was used in the development of that line, otherwise use 'No'. 
5. Show the first 10 rows of lines.

```r
peanut_lines$Bailey_Parent <- grepl("Bailey", peanut_lines$Parentage)
peanut_lines[which(peanut_lines$Bailey_Parent == 1), "Bailey_Parent"] <- "Yes"
peanut_lines[which(peanut_lines$Bailey_Parent == 0), "Bailey_Parent"] <- "No"
peanut_lines[1:10,]
```

6. What percentage of lines were developed using 'Bailey' are verified as high-oleic (Fatty Acid Genotype [FAG] = 'ol ol')? Round your answer to the tenth. Do not include the 'Check' lines.
    
    Note: Bailey (FAG) = '+ +', FAG = 'Check', means it hasn't been identified as high-oleic yet.

```r
round(nrow(peanut_lines[which(peanut_lines$Bailey_Parent=="Yes" & peanut_lines$FAG=="ol ol"),])/nrow(peanut_lines[which(peanut_lines$FAG!="Check"),])*100, 1)
```

7. Replace the name 'Bailey' with the tested name 'N03081T' permanently for all instances in the 'Parentage' column.
8. Show the first 10 lines

```r
peanut_lines$Parentage <- gsub("Bailey", "N03081T", peanut_lines$Parentage)
peanut_lines[1:10,]
```

9. Load the 'peanut_yield.txt' dataset as 'yields'
10. Merge (right join, all.y=TRUE) 'lines' and 'yields' together on 'NC_Accession' and name it 'peanut_db'

```r
yields <- read.table("peanut_yield.txt", header=TRUE, sep="\t")
head(yields)
```

```r
peanut_db <- merge(peanut_lines, yields, by = "NC_Accession", all.y = TRUE)
peanut_db
```

11. Create a dataframe that contains the average yield of Bailey and the average of all other lines with Bailey as the parent.

```r
m <- matrix(nrow = 2, ncol = 1)
df <- data.frame(m)
rownames(df) <- c("Bailey", "Other")
colnames(df) <- "Yield"
df["Bailey",] <- mean(peanut_db[which(peanut_db$NC_Accession=="Bailey"), "Yield"])
df["Other",] <- mean(peanut_db[which(peanut_db$Bailey_Parent=="Yes" & peanut_db$NC_Accession!="Bailey"), "Yield"])
df
```
